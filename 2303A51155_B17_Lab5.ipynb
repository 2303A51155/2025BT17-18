```python
import pandas as pd

students = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 5],
    'name': ['Amit', 'Neha', 'Rahul', 'Pooja', 'Arjun'],
    'gender': ['M', 'F', 'M', 'F', 'M']
})

exam_scores = pd.DataFrame({
    'student_id': [1, 2, 4, 5, 6],
    'math_score': [88, 75, 90, 95, 60],
    'english_score': [92, 85, 87, 90, 70]
})

print("Students Dataset:")
print(students, '\n')

print("Exam Scores Dataset:")
print(exam_scores, '\n')

inner_join = pd.merge(students, exam_scores, on='student_id', how='inner')
print("Inner Join Result:")
print(inner_join, '\n')

left_join = pd.merge(students, exam_scores, on='student_id', how='left')
print("Left Join Result:")
print(left_join, '\n')

right_join = pd.merge(students, exam_scores, on='student_id', how='right')
print("Right Join Result:")
print(right_join, '\n')

print("Aggregate Statistics:")
print("Mean scores:\n", inner_join[['math_score', 'english_score']].mean())
print("\nMax scores:\n", inner_join[['math_score', 'english_score']].max())
print("\nMin scores:\n", inner_join[['math_score', 'english_score']].min())
print("\nCount:\n", inner_join[['math_score', 'english_score']].count())

grouped_stats = inner_join.groupby('gender')[['math_score', 'english_score']].agg(['mean', 'max', 'min', 'count'])
print("\nGrouped Statistics by Gender:")
print(grouped_stats)
```

```python
import pandas as pd

customers = pd.read_csv("customers.csv")
sales = pd.read_csv("sales.csv")

inner_merged = pd.merge(sales, customers, on="customer_id", how="inner")
left_merged = pd.merge(sales, customers, on="customer_id", how="left")

total_sales = inner_merged["amount"].sum()
mean_sales = inner_merged["amount"].mean()
sales_per_customer = inner_merged.groupby("customer_id")["amount"].sum().reset_index()
region_stats = inner_merged.groupby("region")["amount"].agg(["sum", "mean", "count"]).reset_index()

print("Inner Merge:\n", inner_merged)
print("\nLeft Merge:\n", left_merged)
print("\nTotal Sales:", total_sales)
print("Mean Sales:", mean_sales)
print("\nSales per Customer:\n", sales_per_customer)
print("\nRegion-wise Stats:\n", region_stats)
```
